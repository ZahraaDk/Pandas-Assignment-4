After doing an initial analysis on the prices, i will apply the correlation function to get some additional insights on the relationship between each attribute and the target(price). The correlation function gives a values between 1(Strong positive correlation) to -1(Strong negative correlation) with 0 correlation meaning no relationship. It reflects how similar the measurements of variables are across a dataset and identifies patterns and relationships that exist within the data. However, the default .corr() function (which uses pearson's method) is usually applied on numeric variables.  

```python
correlation = df.corr()
price_corr = correlation['price'].sort_values(ascending=False)
print(price_corr)
```

The correlation method provides us with some good insights; there is a strong correlation between the price and the square foot of living space. However concerning the different variable we have; categorical variables and geographical variables, there isn't much to inspect through this method and it doesnt provide good values on such correlation. The df.corr() method works on continuous numerical variables, but not for ranked/geographic variables. 

```python
correlation = df.corr(method='kendall')
price_corr = correlation['price'].sort_values(ascending=False)
print(price_corr)
```

This correlation shows that the categorical variable; bathrooms has a moderate correlation with the price, however the bedrooms, floors, and view show much weaker positive correlation implying that they may influence the price but to a lesser extent.

```python
df[['price', 'bedrooms', 'bathrooms', 'sqft_living']].groupby(['sqft_living']).mean().round().reset_index()
```

```python
df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'grade']].groupby(['grade']).mean().round().reset_index()
```

There is a proportional relationship between the grade and the price. As the grade increases so does the average price per each grade. Also, higher grades tend to have more sqft living, bathrooms and bedrooms, which would imply that higher grades have larger and more luxurious houses. ```python
df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'grade', 'view']].groupby(['view']).mean().round(1).reset_index()
```

As the view level increase, so does the price. Meaning that properties of higher view levels (better views) often have higher prices. The relationship between the number of floors and the price of the property is not always straightforward, where there are cases where the increase in the number of floors doesn't necessarily imply an increase in the price. We can deduce from this analysis that one factor may not necessarily affect the target at its own, however it may do so along with various factors, that together may give an increased influence on the target. ```python
df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'grade', 'view', 'waterfront']].groupby(['waterfront']).mean().round().reset_index()
```

Through the above dataframes, we can get many insights.
